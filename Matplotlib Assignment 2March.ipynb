## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of numerical data through their quartiles. The box plot consists of a rectangular box, which represents the middle 50% of the data, and two lines, called whiskers, which extend from the box and represent the minimum and maximum values within a specified range.

```python
data = [box1, box2]

fig, ax = plt.subplots()

ax.boxplot(data)

ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box plot of Box 1 and Box 2')

plt.show()
```
