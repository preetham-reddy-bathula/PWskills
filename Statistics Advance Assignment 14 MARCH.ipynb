```python
#Q1
import scipy.stats as stats
import numpy as np

def f_test(x, y):
    x = np.array(x)
    y = np.array(y)
    f = np.var(x, ddof=1)/np.var(y, ddof=1) # calculate F test statistic 
    dfn = x.size-1 # degrees of freedom numerator 
    dfd = y.size-1 # degrees of freedom denominator 
    p = 1-stats.f.cdf(f, dfn, dfd) # find p-value of F test statistic 
    return f, p
```

```python
#Q2
def f_critical(alpha, dfn, dfd):
    lower_crit = stats.f.ppf(alpha/2, dfn, dfd)
    upper_crit = stats.f.ppf(1-alpha/2, dfn, dfd)
    return lower_crit, upper_crit
```

```python
#Q3
def f_test_random_samples(mu1, mu2, sigma1, sigma2, size):
    x = np.random.normal(mu1, sigma1, size)
    y = np.random.normal(mu2, sigma2, size)
    return f_test(x, y)
```

```python
#Q4
def f_test_known_variances(var1, var2, n1, n2, alpha):
    f = var1 / var2
    dfn = n1 - 1
    dfd = n2 - 1
    lower_crit, upper_crit = f_critical(alpha, dfn, dfd)
    return f, lower_crit, upper_crit
```

```python
#Q5
def f_test_manufacturer_claim(claimed_var, sample_var, n, alpha):
    return f_test_known_variances(sample_var, claimed_var, n, 1, alpha)
```

```python
#Q6
def f_distribution_properties(dfn, dfd):
    mean = dfd / (dfd - 2) if dfd > 2 else None
    var = (2 * dfd * dfd * (dfn + dfd - 2)) / (dfn * (dfd - 2) * (dfd - 2) * (dfd - 4)) if dfd > 4 else None
    return mean, var
```

```python
#Q7
def f_test_normal_samples(var1, n1, var2, n2, alpha):
    return f_test_known_variances(var1, var2, n1, n2, alpha)
```

```python
#Q8
def f_test_restaurant_times(times_a, times_b, alpha):
    return f_test(times_a, times_b)
```

```python
#Q9
def f_test_student_scores(scores_a, scores_b, alpha):
    return f_test(scores_a, scores_b)
```
