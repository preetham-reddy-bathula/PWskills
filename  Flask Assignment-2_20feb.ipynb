## Q1. Explain GET and POST methods.

GET and POST are HTTP methods used for sending data between the client and the server.

GET method is used to retrieve data from the server. It sends data in the form of a query string appended to the URL. This method is used when there is no need to send sensitive information, and data can be cached.

POST method is used to submit data to the server. The data is sent in the request body, which is not visible in the URL. This method is used when there is a need to send sensitive information and data cannot be cached.

## Q2. Why is request used in Flask?

The request object in Flask is used to handle incoming client requests. It contains information about the request made by the client, such as the HTTP method used, the data submitted, the headers, and the URL. The request object is used to extract information from the request, which can be used to process the request and generate a response.

## Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect the client to a different URL. It is commonly used when a user submits a form or logs in and needs to be redirected to a different page. It is used to improve the user experience by providing a smooth flow of navigation between pages.

## Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are used to generate dynamic HTML pages. They allow you to separate the presentation of the data from the business logic of the application. Templates in Flask are written in HTML and use placeholders to insert dynamic data.

The render_template() function in Flask is used to render the HTML templates. It takes the name of the template file as the argument and renders the template with the dynamic data provided. It allows you to dynamically generate HTML pages based on the user's request.
