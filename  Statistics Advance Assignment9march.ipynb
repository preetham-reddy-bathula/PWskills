Q1: The Probability Mass Function (PMF) is a function that describes the probability of discrete random variables taking certain values. It gives the probability of each possible outcome of a discrete random variable. For example, the PMF of rolling a fair six-sided die gives the probability of getting each number from 1 to 6.

The Probability Density Function (PDF) is a function that describes the probability of continuous random variables taking certain values. It gives the relative likelihood of each possible outcome, rather than the probability of each outcome. For example, the PDF of the height of adult humans in a population gives the relative likelihood of someone being a certain height.

Q2: The Cumulative Density Function (CDF) is a function that gives the probability that a random variable takes a value less than or equal to a certain value. It gives the cumulative probability distribution of a random variable. For example, the CDF of rolling a fair six-sided die gives the probability of getting a number less than or equal to each possible number.

The CDF is used because it allows us to easily calculate the probability of a random variable taking a value within a certain range, by subtracting the probability of the lower end of the range from the probability of the upper end of the range.

Q9: The Central Limit Theorem (CLT) is a statistical theorem that states that the sample means of a large number of independent and identically distributed random variables will be approximately normally distributed, regardless of the distribution of the original variables. This means that even if the original data is not normally distributed, the sample means will be normally distributed if the sample size is sufficiently large.

The significance of the Central Limit Theorem is that it allows us to make inferences about a population based on a sample, even if we do not know the distribution of the population. It also forms the basis for many statistical tests and models.
