Q1: Estimation statistics is a branch of statistics that deals with estimating population parameters from sample statistics. Point estimation involves using a single value to estimate the value of an unknown population parameter, such as using the sample mean to estimate the population mean. Interval estimation involves using a range of values to estimate the value of an unknown population parameter, such as using a confidence interval to estimate the population mean.

Q2: Here's an example Python function to estimate the population mean using a sample mean and standard deviation:

```python
def estimate_population_mean(sample_mean, sample_std, sample_size):
    # Calculate the standard error
    std_error = sample_std / (sample_size ** 0.5)
    # Calculate the margin of error for a 95% confidence interval
    margin_of_error = 1.96 * std_error
    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    # Return the confidence interval
    return (lower_bound, upper_bound)
```

Q3: Hypothesis testing is a statistical method used to determine whether an assumption about a population parameter is supported by the sample data. It is used to test a claim or hypothesis about the population parameter using sample statistics. The importance of hypothesis testing lies in its ability to make inferences about the population from sample data, and to make decisions based on the evidence provided by the sample.

Q4: A hypothesis that states whether the average weight of male college students is greater than the average weight of female college students could be: "The average weight of male college students is greater than the average weight of female college students."

Q5: Here's an example Python script to conduct a hypothesis test on the difference between two population means:

```python
import numpy as np
from scipy.stats import ttest_ind

# Generate two samples from two populations with different means
sample1 = np.random.normal(10, 2, 100)
sample2 = np.random.normal(12, 2, 100)

# Conduct a two-sample t-test
t_stat, p_value = ttest_ind(sample1, sample2)

# Print the results
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

Q13. To estimate the population mean revenue with a 95% confidence interval:

```python
sample_mean = 500
sample_sd = 50
sample_size = 50
confidence_level = 0.95

alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha/2, sample_size - 1)
margin_of_error = t_critical * (sample_sd / np.sqrt(sample_size))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

confidence_interval = (lower_bound, upper_bound)
print("95% Confidence Interval:", confidence_interval)
```

Q14. To test the hypothesis with a significance level of 0.05:

```python
population_mean = 10
sample_mean = 8
sample_sd = 3
sample_size = 100
alpha = 0.05

t_statistic = (sample_mean - population_mean) / (sample_sd / np.sqrt(sample_size))
df = sample_size - 1
p_value = 1 - stats.t.cdf(t_statistic, df)

reject_null = p_value < alpha
print("t-statistic:", t_statistic)
print("p-value:", p_value)
print("Reject null hypothesis:", reject_null)
```
