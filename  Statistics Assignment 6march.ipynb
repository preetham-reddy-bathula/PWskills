For example, eye color is an example of qualitative data, while weight is an example of quantitative data. Continuous data is measured on a scale that can take on any value, such as height or temperature, while discrete data is measured on a scale that can only take on certain values, such as the number of siblings someone has.

## Answer 5

Levels of measurement refer to the different ways that data can be measured, and include nominal, ordinal, interval, and ratio levels. Nominal data is categorical and has no order or rank, such as gender or race. Ordinal data has a natural order or ranking, such as educational level or income level. Interval data has equal intervals between values but no true zero point, such as temperature measured in Celsius or Fahrenheit. Ratio data has equal intervals between values and a true zero point, such as height or weight.

## Answer 10

Some common measures of central tendency include the mean, median, and mode, while common measures of variability include the range, standard deviation, and variance. The mean is the average of all values in the dataset, the median is the middle value when the data is sorted, and the mode is the most common value. The range is the difference between the highest and lowest values, while the standard deviation and variance measure how spread out the data is around the mean. These measures can be used to describe the shape of the distribution and to compare different datasets.
