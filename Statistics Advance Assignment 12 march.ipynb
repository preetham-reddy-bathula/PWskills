```python
#Q1
import scipy.stats as stats

mean = 50
std_dev = 5
n = 30  # assuming sample size of 30, add actual sample size if available
confidence = 0.95

se = std_dev / (n**0.5)
ci = stats.norm.interval(confidence, loc=mean, scale=se)
print(f'The 95% confidence interval is {ci}')
```

Interpretation: The confidence interval represents the range in which we are 95% confident that the true population mean lies.

```python
#Q2
import numpy as np
from scipy.stats import chisquare

observed = np.array([20, 20, 20, 10, 10, 20]) # replace these with your observed values
expected = np.array([20, 20, 20, 10, 10, 20])

chi2, p = chisquare(observed, f_exp=expected)

print(f'Chi-square statistic = {chi2}, p-value = {p}')
```

If the p-value is less than 0.05, we reject the null hypothesis and conclude that the observed and expected distributions are significantly different.

```python
#Q3
from scipy.stats import chi2_contingency

# define the contingency table
data = np.array([[20, 15], [10, 25], [15, 20]])
chi2, p, dof, expected = chi2_contingency(data)

print(f'Chi-square statistic = {chi2}, p-value = {p}')
```

If the p-value is less than 0.05, we reject the null hypothesis and conclude that there is a significant association between the groups and outcomes.

```python
#Q4
p = 60 / 500  # proportion of smokers
n = 500  # total number of individuals

se = (p * (1 - p) / n)**0.5
ci = stats.norm.interval(0.95, loc=p, scale=se)

print(f'The 95% confidence interval for the proportion of smokers is {ci}')
```

```python
#Q6
import matplotlib.pyplot as plt
import numpy as np

df = 10  # degrees of freedom
x = np.linspace(stats.chi2.ppf(0.01, df), stats.chi2.ppf(0.99, df), 100)
plt.plot(x, stats.chi2.pdf(x, df), 'r-', lw=5, alpha=0.6, label='chi2 pdf')
plt.fill_between(x, stats.chi2.pdf(x, df), where=(x>=15), color='grey', alpha=0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Chi-Square Distribution with 10 degrees of freedom')
plt.grid()
plt.show()
```

```python
#q11
import scipy.stats as stats

# given values
n = 30
sample_mean = 72
sample_std_dev = 10
pop_mean = 70

# calculate standard error
se = sample_std_dev / (n**0.5)

# calculate t statistic
t_stat = (sample_mean - pop_mean) / se

# calculate degrees of freedom
df = n - 1

# calculate p-value (two-tailed)
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

print(f'T-statistic = {t_stat}, p-value = {p_value}')
```

This will output the t-statistic and the p-value. If the p-value is less than the significance level of 0.05, you would reject the null hypothesis and conclude that the population mean is significantly different from 70. If the p-value is greater than 0.05, you would fail to reject the null hypothesis and conclude that there is not enough evidence to suggest that the population mean is significantly different from 70.
